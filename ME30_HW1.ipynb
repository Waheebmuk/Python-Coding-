#### Sin Waves vs. Time Graph Code
I was to run some Python code in order to create the desired graph.

```python
from matplotlib import pyplot as plt 
import numpy as np
x = np.linspace(0, 5. * np.pi, 1000) 
y = np.sin(x)
plt.plot(x, y) 
plt.plot(x, y + 2) 
plt.grid(True)
plt.xlabel("Time, s") 
plt.ylabel("Voltage, V")
plt.title("Sine Waves vs. Time")
plt.savefig("images/01_sinewave_plot.svg")
plt.savefig("images/01_sinewave_plot.png")
```

This is because the .svg file is a vector image file format which means it stores the image as a text based file (XML) thus giving instructions on how to set the image. Whereas the .png file is a bitmap image which means the image is made up from a fixed number of pixels. That is why the .png file became blurry because when asked to go over the number of fized pixels it simply zoomed in. On the other hand the .svg file is essentially a blueprint for the image where whatever dimensions you give the image it will create the image at that desired size.
